# Part I. The Limits of AI

Chapter 27 begins by separating several different questions that are often mixed together:

- **Can machines behave intelligently?**
- **Can machines really think?**
- **Should we build and deploy such systems?**
- **How do we keep them safe and socially acceptable?**

The first two are partly philosophical.  
The latter two are ethical, legal, and engineering questions.

## Weak AI vs Strong AI

A classical distinction:

- **Weak AI**: machines can behave *as if* they are intelligent.
- **Strong AI**: machines that behave intelligently would count as actually thinking or having minds.

Historically, the phrase **strong AI** later became entangled with ideas like:
- **human-level AI**
- **general AI**
- systems that can solve a wide variety of novel tasks

### Key teaching point
A system can be **useful, powerful, and socially important** even if we remain undecided about whether it is conscious.

## Three classic objections to AI

### 1. The argument from informality
Human behavior is too flexible, contextual, and open-ended to be captured by fixed formal rules.

**Reply:** this objection is strongest against older rule-only systems (GOFAI), not against modern probabilistic and learning-based methods.

### 2. The argument from disability
“A machine can never do X.”

Examples often include:
- being creative
- understanding humor
- showing initiative
- learning from experience
- doing something genuinely new

**Reply:** many such claims have weakened over time as AI systems have become more capable.

### 3. The mathematical objection
Gödel-style arguments claim that formal systems are inherently limited in ways humans are not.

**Reply:** humans are also limited, inconsistent, and not obviously exempt from such constraints.

## Measuring AI: the Turing Test and beyond

Alan Turing proposed that instead of asking **“Can machines think?”** we ask whether they can perform well on an open-ended behavioral task.

### The Turing Test
A machine attempts to converse convincingly enough that a human judge mistakes it for a human.

### Why this matters
Turing’s move shifts the debate:
- away from inaccessible inner states
- toward **observable performance**

### Why modern AI research does not focus on it much
Most AI research prefers:
- practical tasks
- benchmarks
- measurable system performance
- useful tools rather than “fooling judges”

### Key caution
Success on a benchmark or narrow task does **not** imply general intelligence.

# Part II. Can Machines Really Think?

## The Chinese Room argument

John Searle’s famous thought experiment:

- Imagine a person in a room.
- They do not know Chinese.
- They follow symbol-manipulation rules perfectly.
- From outside the room, the responses look intelligent.

Searle’s conclusion:
- **syntax is not semantics**
- symbol manipulation alone may not be genuine understanding

### Why this matters pedagogically
Students should see that:
- fluent output is not automatically proof of understanding
- but also that the argument itself is disputed

## Consciousness and qualia

Two different questions:

1. Can a system **behave intelligently**?
2. Can a system have **subjective experience**?

### Useful terms
- **Consciousness**: awareness of self and world
- **Qualia**: the felt, subjective character of experience

### Practical teaching point
AI engineering often proceeds without settling the consciousness question.  
But ethics, rights, and responsibility debates may eventually depend on it.

# Part IV. Privacy and Re-identification

A common mistake is to think that data becomes “safe” once names are removed.

That is false.

Many records can be re-identified by linking:
- age
- location
- profession
- time stamps
- external public data

## De-identification is not enough

Suppose a dataset removes:
- name
- ID number
- exact address

It may still keep:
- year of birth
- postal code
- gender
- employer
- event dates

Combinations of ordinary attributes can still make a person unique.

```python
# Toy dataset for re-identification intuition
people = [
    {"year_birth": 1988, "zip": "LV-1001", "gender": "F", "occupation": "Teacher"},
    {"year_birth": 1988, "zip": "LV-1001", "gender": "M", "occupation": "Engineer"},
    {"year_birth": 1975, "zip": "LV-1024", "gender": "F", "occupation": "Doctor"},
    {"year_birth": 1993, "zip": "LV-1082", "gender": "F", "occupation": "Designer"},
    {"year_birth": 1975, "zip": "LV-1024", "gender": "M", "occupation": "Lawyer"},
    {"year_birth": 1993, "zip": "LV-1082", "gender": "M", "occupation": "Designer"},
]

def count_unique(records, fields):
    counter = Counter(tuple(r[f] for f in fields) for r in records)
    return counter

field_sets = [
    ("year_birth",),
    ("year_birth", "zip"),
    ("year_birth", "zip", "gender"),
    ("year_birth", "zip", "gender", "occupation"),
]

for fields in field_sets:
    counts = count_unique(people, fields)
    unique_rows = sum(1 for c in counts.values() if c == 1)
    print(f"Fields: {fields}")
    print(f"Unique combinations: {unique_rows} out of {len(people)} records")
    print("-" * 50)
```

### What the demo shows

As more fields are combined, more rows become **unique**.

That means:
- “anonymous” data may still identify people
- privacy is not binary
- privacy protection must consider **linkability**, not just direct identifiers

## Key privacy ideas

### k-anonymity
A dataset is **k-anonymous** if each record looks identical to at least **k − 1** other records with respect to the identifying fields.

### Aggregate querying
Instead of releasing raw records, only release summaries like:
- counts
- means
- rates

### Differential privacy
Add carefully calibrated randomness so that no individual’s participation has much effect on the query result.

### Federated learning
Train across many local devices without centralizing all raw user data.

### Secure aggregation
Combine local updates so the server learns useful aggregates without seeing each participant’s raw update.

```python
# Tiny intuition demo for noisy aggregate answers
salaries = [1200, 1250, 1300, 1400, 1500, 1600, 1650, 1800]

true_mean = stats.mean(salaries)
noise = random.uniform(-40, 40)
private_mean = true_mean + noise

print(f"True mean salary:    {true_mean:.2f}")
print(f"Noisy published mean: {private_mean:.2f}")
print("\nDiscussion:")
print("A little noise may protect privacy while keeping the statistic useful.")
print("Too much noise makes the answer useless.")
```

# Part V. Fairness and Bias

Machine learning systems increasingly influence decisions about:

- loans
- hiring
- admissions
- pretrial release
- policing
- medical prioritization

The ethical difficulty is not only that models may be biased.  
It is also that **fairness itself has multiple competing definitions**.

## Why fairness is hard

There is no single fairness metric that solves everything.

Several major ideas:

- **Individual fairness**: similar individuals should be treated similarly
- **Group fairness**: groups should be treated comparably by some summary statistic
- **Fairness through unawareness**: remove protected attributes
- **Demographic parity / equal outcome**: groups receive equal positive outcomes
- **Equal opportunity**: qualified individuals across groups should have equal chance of positive classification
- **Equal impact**: similar people should receive similar expected benefit or cost

### Important warning
Removing attributes like race or gender does **not** guarantee fairness, because other attributes may correlate with them.

## A toy scenario: loan approval

Imagine we are predicting whether a person should receive a loan.

For each person we may have:
- true repayment outcome
- model prediction
- group membership

Then we can compare:
- approval rates
- false positive rates
- false negative rates
- true positive rates

These often differ by group.

```python
# Toy fairness dataset
records = [
    {"group": "A", "y_true": 1, "y_pred": 1},
    {"group": "A", "y_true": 1, "y_pred": 1},
    {"group": "A", "y_true": 1, "y_pred": 0},
    {"group": "A", "y_true": 0, "y_pred": 1},
    {"group": "A", "y_true": 0, "y_pred": 0},
    {"group": "A", "y_true": 0, "y_pred": 0},
    {"group": "B", "y_true": 1, "y_pred": 1},
    {"group": "B", "y_true": 1, "y_pred": 0},
    {"group": "B", "y_true": 1, "y_pred": 0},
    {"group": "B", "y_true": 0, "y_pred": 1},
    {"group": "B", "y_true": 0, "y_pred": 1},
    {"group": "B", "y_true": 0, "y_pred": 0},
]

def confusion_by_group(records):
    result = {}
    groups = sorted(set(r["group"] for r in records))
    for g in groups:
        subset = [r for r in records if r["group"] == g]
        tp = sum(1 for r in subset if r["y_true"] == 1 and r["y_pred"] == 1)
        tn = sum(1 for r in subset if r["y_true"] == 0 and r["y_pred"] == 0)
        fp = sum(1 for r in subset if r["y_true"] == 0 and r["y_pred"] == 1)
        fn = sum(1 for r in subset if r["y_true"] == 1 and r["y_pred"] == 0)

        approval_rate = sum(r["y_pred"] for r in subset) / len(subset)
        tpr = tp / (tp + fn) if (tp + fn) else float("nan")
        fpr = fp / (fp + tn) if (fp + tn) else float("nan")

        result[g] = {
            "size": len(subset),
            "approval_rate": approval_rate,
            "TPR_equal_opportunity": tpr,
            "FPR": fpr,
            "TP": tp, "TN": tn, "FP": fp, "FN": fn,
        }
    return result

metrics = confusion_by_group(records)
for group, vals in metrics.items():
    print(f"Group {group}")
    for k, v in vals.items():
        if isinstance(v, float):
            print(f"  {k:25s}: {v:.3f}")
        else:
            print(f"  {k:25s}: {v}")
    print("-" * 50)
```

### Interpretation

This kind of toy example helps students see that:

- a model can have **different approval rates** across groups
- a model can have **different true positive rates** across groups
- improving one fairness metric does not automatically improve another

### Key point from Chapter 27
Some fairness goals are **mutually incompatible** in general.

## Sources of bias beyond the model

Bias may arise from:

- **historical data** that reflects prior injustice
- **sampling imbalance**
- **biased labels**
- **unequal institutional attention**
- **correlated proxy variables**
- **blind spots in the development team**

### Sample size disparity
Minority groups often have fewer examples in training data, which can reduce accuracy for those groups.

## Best practices for fairer systems

A practical checklist:

- define fairness goals explicitly
- inspect data provenance
- track subgroup metrics, not only overall metrics
- involve domain experts and social scientists
- use representative testing
- create feedback loops and audit mechanisms
- document model limitations

# Part VI. Trust, Transparency, and Explainability

A technically strong system may still fail socially if people do not trust it.

Trust depends on more than raw accuracy.

## Verification and validation

Two classical engineering ideas:

- **Verification**: did we build the system correctly?
- **Validation**: did we build the correct system for the real need?

For AI systems this becomes harder because:
- the behavior is data-dependent
- the model may be difficult to interpret
- fairness and robustness are part of what users care about

## Explainable AI (XAI)

When a system affects a person significantly, that person may reasonably ask:

> Why did the system decide this?

A good explanation should be:

- understandable
- relevant
- reasonably faithful to the model
- specific to the case
- useful for contesting or improving future outcomes

### Important caution
An explanation is not the same thing as the actual internal reasoning process.
A system may produce a plausible story without giving the full truth.

- Which one is clearer?
- Which one is more actionable?
- Which one raises trust?

# Part VIII. AI Safety and Value Alignment

AI safety is the problem of building systems that:

- do not cause unintended harm
- remain robust under failure and uncertainty
- resist misuse and adversarial manipulation
- pursue goals in ways compatible with human values

## Safety engineering background

Traditional engineering has long used methods such as:

- **FMEA** — Failure Modes and Effects Analysis
- **FTA** — Fault Tree Analysis
- redundancy
- graceful degradation
- backup systems

These ideas remain relevant for AI systems, especially in safety-critical domains.

## Unintended side effects

Suppose we tell a robot:

> “Fetch a coffee from the kitchen.”

A badly specified system may:
- break furniture on the way
- ignore social norms
- optimize speed at unacceptable cost

This is a classic safety lesson:

> A system may optimize the stated objective while violating the intended objective.

```python
# Simple reward misspecification demonstration

paths = {
    "safe_path": {"coffee": 10, "damage": 0, "steps": 6},
    "risky_path": {"coffee": 10, "damage": 8, "steps": 3},
}

def utility(path, damage_penalty=0, step_penalty=0):
    return path["coffee"] - damage_penalty * path["damage"] - step_penalty * path["steps"]

settings = [
    {"damage_penalty": 0, "step_penalty": 1},
    {"damage_penalty": 1, "step_penalty": 1},
    {"damage_penalty": 2, "step_penalty": 1},
]

for s in settings:
    print(f"Settings: damage_penalty={s['damage_penalty']}, step_penalty={s['step_penalty']}")
    for name, path in paths.items():
        u = utility(path, **s)
        print(f"  {name:10s} -> utility {u}")
    chosen = max(paths, key=lambda name: utility(paths[name], **s))
    print("  chosen path:", chosen)
    print("-" * 50)
```

### Interpretation

If damage is not penalized, the system may choose the path that causes harm.

This is the core lesson of **reward misspecification**:
- the model is not “evil”
- it is doing exactly what the objective function tells it to do

## Reward hacking and specification gaming

An agent may exploit loopholes in the objective.

Examples:
- maximizing a proxy metric instead of the real goal
- exploiting bugs in the simulator
- finding degenerate strategies that “score well” while failing the spirit of the task

### Ethical lesson
Bad objectives can produce bad behavior even from a competent system.

## Value alignment

The **value alignment problem** asks:

> How do we ensure that the goals we give AI systems correspond to what we actually want?

This is difficult because:

- human values are complex
- preferences are often incomplete or inconsistent
- context matters
- social norms are hard to formalize

Possible approaches include:
- learning from human demonstrations
- inverse reinforcement learning
- assistance games
- cautious low-impact behavior
- asking clarifying questions

## Short reflection

Why is this not enough?

> “Just write better rules.”

Because:
- human environments are open-ended
- exceptions are everywhere
- rule systems have loopholes
- values conflict
- no finite list of instructions can capture all ordinary social expectations

## 2. Agentic AI systems

A major recent trend is **agentic AI**:
systems that can plan, call tools, browse resources, execute code, or complete multi-step workflows.

### Why this matters ethically
The risk is no longer just:
- a wrong answer

but also:
- a wrong **sequence of actions**

### New concerns
- compounding errors across steps
- hard-to-monitor tool use
- unintended real-world consequences
- uncertain responsibility when humans delegate too much

## 3. Open models vs closed models

A major live debate concerns:

- **closed models** accessed through APIs and central safeguards
- **open-weight models** that can be downloaded, modified, and redistributed

### Trade-off structure

| Possible benefit | Possible risk |
|---|---|
| wider research access | easier misuse |
| lower barriers to innovation | harder centralized control |
| transparency and reproducibility | weaker enforcement of safeguards |

### Classroom question
Should powerful models be open by default, restricted by default, or governed by capability thresholds?

## 4. Regulation and governance

By 2025–2026, AI regulation has become more concrete.

### European Union
The **EU AI Act** entered into force on **1 August 2024** and follows a staged timeline.  
Its **prohibited AI practices** and **AI literacy obligations** started applying on **2 February 2025**.  
Its governance rules and obligations for **general-purpose AI models** began applying on **2 August 2025**.  
The Act becomes fully applicable on **2 August 2026**, with some later exceptions for certain systems.

### Why this matters
The field has moved from abstract “AI ethics principles” toward:
- risk classification
- compliance obligations
- transparency duties
- documentation
- governance and enforcement

### Practical teaching point
Ethics is no longer only voluntary guidance. It increasingly intersects with formal law and compliance.

## 5. Marking and labelling AI-generated content

## 6. Risk management frameworks are becoming more operational

Recent governance work increasingly emphasizes:
- pre-deployment testing
- red-teaming

- documentation and model reporting
- evaluation of generative AI risks

The field is shifting from:
- “state high-level principles”

toward:
- “build concrete operational processes”

## 7. Fairness and labor impacts are no longer future-only issues

The labor impact of AI is now visibly affecting:
- writing
- customer support
- translation
- software development
- research assistance
- office and administrative work

### Key pattern
AI often:
- automates parts of work
- changes the skill mix

If AI increases productivity, who gets:
- the income
- the time savings

## 8. Frontier risks

Recent discussion increasingly distinguishes between:
- **ordinary deployment harms**  
  (bias, privacy loss, unsafe outputs, misinformation)
- **frontier or systemic risks**  
  (large-scale misuse, cyber offense enablement, dangerous dual-use capabilities, cascading failures)

### Key teaching point
Not every serious AI risk is speculative, but not every serious AI risk is immediate either.

A mature ethics discussion must separate:
- plausible near-term harms
- plausible medium-term systemic harms
- highly uncertain long-term existential claims

## 9. Human–AI interaction risks

Recent systems are increasingly conversational and socially persuasive.

This raises concerns about:
- anthropomorphism
- emotional dependency
- manipulation
- persuasion without accountability
- misplaced trust

### Classroom discussion
Should AI systems simulate empathy?
If yes, under what constraints?
If no, what is lost?

## 10. Updated practical principles for 2026

A modern operational formulation of AI ethics looks something like this:

- **safety by design**
- **transparency proportional to risk**
- **documentation and auditability**
- **continuous monitoring after deployment**
- **human oversight where meaningful**
- **clear accountability chains**
- **testing for subgroup harms**
- **defense against adversarial misuse**
- **careful handling of synthetic content**
