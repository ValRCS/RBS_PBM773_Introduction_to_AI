## A friendly one-sentence definition

> Deep learning is a way of training computers to recognize patterns and make decisions by learning directly from large amounts of data, using layered networks that gradually build understanding from simple features to complex ideas.

## Deep Learning — Key Ideas (Approachable Outline)

### 1. From Rules to Learning from Data
- Traditional programming:
  - Humans write rules
- Machine learning:
  - Systems learn rules from examples

**Core shift:** instead of telling the computer exactly *how* to solve a problem, we show it examples and let it discover patterns.

**Simple example:**  
Spam detection can be built either by writing rules such as “contains FREE” or by learning from thousands of labeled emails.

### 2. What Is a Neural Network (Intuition Only)
A neural network is made of layers of small computational units. Each unit:
- takes inputs,
- applies a transformation,
- passes the result forward.

A good mental model is a **chain of filters**, where each layer refines the signal.

### 3. Learning as Trial-and-Error (Training)
The system:
1. makes a prediction,
2. compares it to the correct answer,
3. adjusts itself slightly.

This repeats many times.

**Key idea:** learning is many small improvements over many examples.

### 4. Layers Build Understanding Step by Step
- Early layers: simple patterns such as edges, sounds, or letter fragments
- Middle layers: combinations such as shapes, syllables, or word patterns
- Later layers: higher-level meaning such as objects, topics, or intent

For images: **pixels → edges → shapes → objects**.

### 5. Why “Deep” Matters
“Deep” means the network has multiple layers. More layers allow the model to build more complex internal representations.

A useful analogy is language:
**letters → words → sentences → meaning**.

### 6. Data Is the Fuel
Deep learning works best with lots of data. Usually:
- more data helps,
- more computation helps,
- larger models can help.

These systems learn from experience rather than explicit instructions.

### 7. Generalization: Learning Beyond Examples
The goal is not just to memorize the training set. The goal is to perform well on **new unseen data**.

Example: a model should recognize a new dog photo, not only the dog photos it already saw.

### 8. Different Architectures for Different Problems
Different neural network structures suit different data types.

- **CNNs** are good for images
- **RNNs** and especially **Transformers** are good for sequences such as text or speech

**Key idea:** the architecture should reflect the structure of the problem.

### 9. Strengths of Deep Learning
Deep learning is powerful because it:
- handles messy real-world data,
- learns useful features automatically,
- scales well with data and compute,
- powers many modern AI systems.

### 10. Limitations and Trade-offs
Deep learning also has costs:
- it can require a lot of data,
- it can require substantial compute,
- it may be difficult to interpret,
- it can reproduce biases present in the data.

### 11. Where You See It Today
Examples include:
- voice assistants,
- machine translation,
- recommendation systems,
- medical image analysis,
- autonomous systems,
- large language models.

### Bridge forward
Now that the intuition is in place, the next step is to open the box and see how a single neuron and a simple network work mathematically.

## Mathematical foundations

### 1. The simplest unit: a single neuron

A neuron computes

\[
y = \sigma(w_1 x_1 + w_2 x_2 + \cdots + w_n x_n + b)
\]

where:

- \(x_i\): inputs
- \(w_i\): weights
- \(b\): bias
- \(\sigma\): activation function
- \(y\): output

A neuron first forms a weighted sum and then applies a nonlinearity.

### 2. Vector form

Instead of writing every term separately, we can write:

\[
y = \sigma(\mathbf{w} \cdot \mathbf{x} + b)
\]

This is a compact way to express the same idea.

### 3. Activation functions

Without an activation function, the model is only linear.

Two common activations are:

**Sigmoid**
\[
\sigma(z) = \frac{1}{1+e^{-z}}
\]

**ReLU**
\[
\mathrm{ReLU}(z) = \max(0, z)
\]

Nonlinearity is what allows networks to learn more complex patterns.

### 4. From one neuron to a layer

A whole layer can be written as:

\[
\mathbf{y} = \sigma(W\mathbf{x} + \mathbf{b})
\]

where \(W\) is a weight matrix and \(\mathbf{b}\) is a bias vector.

### 5. Building a network

A neural network is a composition of layers:

\[
\mathbf{y} = f_L(f_{L-1}(\dots f_1(\mathbf{x})))
\]

This compositional structure is the mathematical reason deep networks can represent complex functions.

### 6. Supervised learning setup

We assume training examples of the form \((x, y)\), where:
- \(x\) is the input,
- \(y\) is the correct output.

The model produces a prediction:

\[
\hat{y} = f(x; \theta)
\]

where \(\theta\) is the set of parameters.

### 7. Loss function

A loss function measures how wrong the prediction is.

For regression, one common example is mean squared error:

\[
\mathcal{L} = (y - \hat{y})^2
\]

For binary classification, a more common choice is **binary cross-entropy**.

### 8. Optimization objective

Training means choosing parameters that minimize the loss:

\[
\min_{\theta} \mathcal{L}(f(x; \theta), y)
\]

### 9. Gradient descent

The basic update rule is:

\[
\theta \leftarrow \theta - \eta \nabla_\theta \mathcal{L}
\]

where:
- \(\eta\) is the learning rate,
- \(\nabla_\theta \mathcal{L}\) is the gradient of the loss with respect to the parameters.

### 10. Backpropagation

Because the network is a composition of functions, we can efficiently compute gradients using the chain rule.

That procedure is called **backpropagation**.

### 11. Training loop

A minimal training cycle looks like this:

1. Forward pass: compute predictions  
2. Compute loss  
3. Backward pass: compute gradients  
4. Update parameters  

Repeat over many examples or batches.

### Minimal summary

- Neuron:
  \[
  y = \sigma(\mathbf{w}\cdot\mathbf{x}+b)
  \]
- Layer:
  \[
  \mathbf{y} = \sigma(W\mathbf{x}+\mathbf{b})
  \]
- Loss:
  \[
  \mathcal{L}(y,\hat{y})
  \]
- Update:
  \[
  \theta \leftarrow \theta - \eta \nabla_\theta \mathcal{L}
  \]

## Example 1 — a single neuron for binary classification

We now implement the simplest practical binary classifier: a **single neuron with sigmoid activation**.

Its prediction rule is:

\[
\hat{y} = \sigma(\mathbf{w}\cdot\mathbf{x}+b)
\]

This is mathematically equivalent to logistic regression.

We will train it on the **OR** truth table, which is linearly separable.

````python
# OR dataset
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
], dtype=float)

y = np.array([[0], [1], [1], [1]], dtype=float)

np.random.seed(42)
w = np.random.randn(2, 1)
b = np.zeros((1,))

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def forward_single(X, w, b):
    return sigmoid(X @ w + b)

def binary_cross_entropy(y_true, y_pred):
    eps = 1e-8
    return -np.mean(
        y_true * np.log(y_pred + eps) +
        (1 - y_true) * np.log(1 - y_pred + eps)
    )

learning_rate = 0.1
epochs = 2000

loss_history = []

for epoch in range(epochs):
    y_hat = forward_single(X, w, b)

    m = X.shape[0]
    dz = y_hat - y
    dw = (X.T @ dz) / m
    db = np.sum(dz) / m

    w -= learning_rate * dw
    b -= learning_rate * db

    loss = binary_cross_entropy(y, y_hat)
    loss_history.append(loss)

    if epoch % 400 == 0:
        print(f"Epoch {epoch:4d} | Loss: {loss:.6f}")

print("\nFinal probabilities:")
print(np.round(forward_single(X, w, b), 4))

print("\nPredicted classes:")
print((forward_single(X, w, b) > 0.5).astype(int))

print("\nTrue classes:")
print(y.astype(int))
````

````python
plt.figure(figsize=(6, 4))
plt.plot(loss_history)
plt.xlabel("Epoch")
plt.ylabel("Binary cross-entropy loss")
plt.title("Training loss for a single-neuron OR classifier")
plt.grid(True)
plt.show()
````

## Visualizing the decision boundary

For a single neuron, the decision boundary is defined by:

\[
w_1 x_1 + w_2 x_2 + b = 0
\]

That is a straight line in two dimensions, so a single neuron can only learn a **linear** boundary.

````python
def plot_single_neuron_boundary(X, y, w, b, title):
    x_min, x_max = -0.5, 1.5
    y_min, y_max = -0.5, 1.5

    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 250),
        np.linspace(y_min, y_max, 250)
    )

    grid = np.c_[xx.ravel(), yy.ravel()]
    probs = forward_single(grid, w, b).reshape(xx.shape)

    plt.figure(figsize=(6, 6))
    plt.contourf(xx, yy, probs, levels=20, alpha=0.6)
    plt.colorbar(label="Predicted probability of class 1")
    plt.contour(xx, yy, probs, levels=[0.5], linewidths=2)

    for i in range(len(X)):
        if y[i, 0] == 0:
            plt.scatter(X[i, 0], X[i, 1], marker='o', s=120, label='Class 0' if i == 0 else "")
        else:
            plt.scatter(X[i, 0], X[i, 1], marker='x', s=120, label='Class 1' if i == 1 else "")

    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

plot_single_neuron_boundary(X, y, w, b, "Decision boundary learned for OR")
````

## Why a single layer fails on XOR

The XOR truth table is:

- \((0,0) \to 0\)
- \((0,1) \to 1\)
- \((1,0) \to 1\)
- \((1,1) \to 0\)

These four points are **not linearly separable**. That means no single straight line can separate the positive and negative examples correctly.

So a single neuron is not enough.

````python
# XOR dataset
X_xor = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
], dtype=float)

y_xor = np.array([[0], [1], [1], [0]], dtype=float)

# Train a single neuron on XOR just to show the limitation
np.random.seed(42)
w_xor = np.random.randn(2, 1)
b_xor = np.zeros((1,))

loss_history_xor_single = []

for epoch in range(3000):
    y_hat = forward_single(X_xor, w_xor, b_xor)

    m = X_xor.shape[0]
    dz = y_hat - y_xor
    dw = (X_xor.T @ dz) / m
    db = np.sum(dz) / m

    w_xor -= 0.1 * dw
    b_xor -= 0.1 * db

    loss_history_xor_single.append(binary_cross_entropy(y_xor, y_hat))

print("Probabilities from single neuron on XOR:")
print(np.round(forward_single(X_xor, w_xor, b_xor), 4))

print("\nPredicted classes:")
print((forward_single(X_xor, w_xor, b_xor) > 0.5).astype(int))

print("\nTrue classes:")
print(y_xor.astype(int))
````

````python
plt.figure(figsize=(6, 4))
plt.plot(loss_history_xor_single)
plt.xlabel("Epoch")
plt.ylabel("Binary cross-entropy loss")
plt.title("Single neuron trained on XOR")
plt.grid(True)
plt.show()

plot_single_neuron_boundary(X_xor, y_xor, w_xor, b_xor, "Single-neuron boundary on XOR (fails)")
````

## Example 2 — adding enough depth to learn XOR

The smallest standard network that can represent XOR has:

- 2 input units
- 1 hidden layer
- 2 hidden neurons
- 1 output neuron

This is often written as **2–2–1**.

The network computes:

\[
h = \sigma(XW_1 + b_1)
\]

\[
\hat{y} = \sigma(hW_2 + b_2)
\]

Because of the hidden layer, the model can form a nonlinear decision boundary.

````python
np.random.seed(42)

input_size = 2
hidden_size = 2
output_size = 1

W1 = np.random.randn(input_size, hidden_size)
b1 = np.zeros((1, hidden_size))

W2 = np.random.randn(hidden_size, output_size)
b2 = np.zeros((1, output_size))

def sigmoid_derivative(a):
    return a * (1 - a)

def forward_two_layer(X, W1, b1, W2, b2):
    z1 = X @ W1 + b1
    a1 = sigmoid(z1)
    z2 = a1 @ W2 + b2
    a2 = sigmoid(z2)
    return z1, a1, z2, a2

learning_rate = 1.0
epochs = 10000
loss_history_xor_deep = []

for epoch in range(epochs):
    z1, a1, z2, y_hat = forward_two_layer(X_xor, W1, b1, W2, b2)

    m = X_xor.shape[0]

    dz2 = y_hat - y_xor
    dW2 = (a1.T @ dz2) / m
    db2 = np.sum(dz2, axis=0, keepdims=True) / m

    dz1 = (dz2 @ W2.T) * sigmoid_derivative(a1)
    dW1 = (X_xor.T @ dz1) / m
    db1 = np.sum(dz1, axis=0, keepdims=True) / m

    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1

    loss = binary_cross_entropy(y_xor, y_hat)
    loss_history_xor_deep.append(loss)

    if epoch % 2000 == 0:
        print(f"Epoch {epoch:5d} | Loss: {loss:.6f}")

_, _, _, y_hat = forward_two_layer(X_xor, W1, b1, W2, b2)

print("\nPredicted probabilities:")
print(np.round(y_hat, 4))

print("\nPredicted classes:")
print((y_hat > 0.5).astype(int))

print("\nTrue classes:")
print(y_xor.astype(int))
````

````python
plt.figure(figsize=(6, 4))
plt.plot(loss_history_xor_deep)
plt.xlabel("Epoch")
plt.ylabel("Binary cross-entropy loss")
plt.title("Two-layer network trained on XOR")
plt.grid(True)
plt.show()
````

````python
def predict_proba_two_layer(grid_points, W1, b1, W2, b2):
    _, _, _, out = forward_two_layer(grid_points, W1, b1, W2, b2)
    return out

x_min, x_max = -0.5, 1.5
y_min, y_max = -0.5, 1.5

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 300),
    np.linspace(y_min, y_max, 300)
)

grid = np.c_[xx.ravel(), yy.ravel()]
Z = predict_proba_two_layer(grid, W1, b1, W2, b2).reshape(xx.shape)

plt.figure(figsize=(6, 6))
plt.contourf(xx, yy, Z, levels=20, alpha=0.6)
plt.colorbar(label="Predicted probability of class 1")
plt.contour(xx, yy, Z, levels=[0.5], linewidths=2)

for i in range(len(X_xor)):
    if y_xor[i, 0] == 0:
        plt.scatter(X_xor[i, 0], X_xor[i, 1], marker='o', s=120, label='Class 0' if i == 0 else "")
    else:
        plt.scatter(X_xor[i, 0], X_xor[i, 1], marker='x', s=120, label='Class 1' if i == 1 else "")

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("XOR learned by a 2–2–1 network")
plt.legend()
plt.grid(True)
plt.show()
````

## Side section — why modern hidden layers usually use ReLU instead of sigmoid

Two common activation functions are:

**Sigmoid**
\[
\sigma(x) = \frac{1}{1 + e^{-x}}
\]

**ReLU**
\[
\mathrm{ReLU}(x) = \max(0, x)
\]

### The main reason: gradient flow

For sigmoid,
\[
\sigma'(x) = \sigma(x)(1-\sigma(x))
\]

Its derivative is always at most \(0.25\), and in the saturated regions it becomes very small.  
In deep networks, multiplying many small derivatives together can make gradients shrink toward zero. This is the **vanishing gradient** problem.

For ReLU,
\[
\mathrm{ReLU}'(x) =
\begin{cases}
1, & x > 0 \\\\
0, & x \le 0
\end{cases}
\]

For positive activations, gradients pass through much more directly.

### Practical consequences

Hidden layers often use ReLU because it:
- trains faster,
- is computationally simpler,
- reduces vanishing-gradient problems,
- often leads to more effective optimization in deep networks.

### Where sigmoid is still useful

Sigmoid is still common in the **output layer** for binary classification, because it maps values into \((0,1)\), which fits probability-style outputs.

### One-sentence takeaway

Modern networks prefer **ReLU in hidden layers** because it helps gradients flow better and therefore makes deep networks easier to train.

````python
def relu(x):
    return np.maximum(0, x)

x = np.linspace(-6, 6, 400)
sig = sigmoid(x)
rel = relu(x)

plt.figure(figsize=(7, 4))
plt.plot(x, sig, label="sigmoid")
plt.plot(x, rel, label="ReLU")
plt.xlabel("x")
plt.ylabel("activation")
plt.title("Sigmoid and ReLU")
plt.grid(True)
plt.legend()
plt.show()
````
